```python
from collections import defaultdict

budgets = defaultdict(int)
budgets[2002] = 500

for year in range(2000, 2005):
    budgets[year] += 100
    
    
print(budgets)
```

```python
def sum(a):
    global b
    print('sum', a, b)
    b += 1
    
def sum_no_override(a):
    print('sum', a, b)


def sum_b5(a):
    b = 5
    print('sum_b5', a, b)
    b += 1

b = 3

def wrapper():
    b = 1
    def sum_nonlocal(a):
        nonlocal b
        print('sum_nonlocal', a, b)
        b += 1
    sum_non_local()


sum(1)
sum_no_override(1)
sum_b5(1)

wrapper()
```
