# Iteration and iterators

https://docs.python.org/3/library/stdtypes.html#iterator-types
https://docs.python.org/3/library/itertools.html

```python
for number in [1,2,3]:
    print(number)
    
print("under the hood it is:")

iterator = iter([1,2,3]) # which is the same as [1,2,3].__iter__()
stop = False
while not stop:
    try:
        number = next(iterator) # which is the same as iterator.__next__()
        print(number)
    except StopIteration:
        stop = True
        
[1,2,3].__iter__()
```

# Helpful syntax


```python
print("unpacking")
a, b, c = [1,2,3]
print(a,b,c)
print("also works with head, tail")
first, (second_1, second_2), *tail = [1, [2,3], 4, 5,6]
print(first, second_1, tail)
```

```python
print("default values")
def sum(a, b=2, should_print=True):
    val = a + b
    if should_print:
        print('{} + {} = {}'.format(a,b,val))
    return val

sum(1)
sum(a=2)
sum(2, b=3)
sum(should_print=False, b=1, a=2)

try:
    sum(b=1)
except Exception as exc:
    print('only problem is when invocation is not clear:', exc)
    

print("also def sum(a=1,b) makes sum(1) be ambiguous , so the definition is not allowed. Optional parameters must be specified after required ones.")
```

# Intention of __*__ protocol methods


```python

print("to write library/glue code like this:")
class Element():
    
    def __init__(self, name):
        # Ignore the __init__ for now
        # self is like this in js
        self.name = name
        self.children = []
        
    def add(self, child):
        self.children.append(child)
        return self
    
    def __iter__(self):
        # specify how to iterate an element
        return iter(self.children) # or you could write your own
    
    def __str__(self):
        children = ''.join(str(child) for child in self.children) # js join, different only in how you invoke it
        return '<{name}>{children}</{name}>'.format(name=self.name, children=children)
    
    def __getitem__(self, item_index):
        return self.children[item_index]
    

body = Element('body')
div = Element('div')
pre = Element('pre')

print('so you can write user code like this:')
body.add(div).add(pre)
print(body)
for element in body:
    print(element)
print(body[0])
```
